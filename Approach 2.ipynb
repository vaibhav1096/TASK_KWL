```python
import pandas as pd
import seaborn as sns
```

### Analyzing Data

```python
df= pd.read_csv('level_progress.csv')
```

```python
a=df.dtypes[df.dtypes=='object'].index
df[a].describe()
```

```python
#sorting only on fail levels and resetting the index
```

```python
a=df[df['status']=='fail'].sort_values('player_id').reset_index()
```

```python
a
```

```python
# groupping By with player ID and level_number
#aim here is to get every possible fail level for each player
```

```python
z=a.groupby(['player_id','level_number'])
```

```python
a=z.describe()
a
```

```python
x=pd.DataFrame(a['index']['count']).reset_index()
```

```python
# sorting for the levels which have high count of failure for a given level for a given player and capturing it in series
```

```python
x.sort_values(['player_id','count'],ascending = (True, False))
```

```python
dit={}
for i in range(len(x)):
    if x['player_id'][i] not in dit:
        dit[x['player_id'][i]]=x['level_number'][i]

        
        
```

```python
#the following series contains the level at which a player faces maximum failure
# mode for all the players will tell us which level dailure is max
```

```python
a=pd.Series(dit)
a
```

```python
import statistics
```

```python
print("most likely failure level {}".format(statistics.mode(pd.Series(dit))))
```
